# Statistics for Machine Learning

This notebook covers basic statistics that are essential for understanding and evaluating machine learning models. Topics include measures of central tendency, dispersion, probability distributions, and hypothesis testing.


### **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Visual settings
sns.set(style="whitegrid")
```

### **Measures of Central Tendency**

```python
data = [12, 15, 12, 18, 19, 12, 14, 20, 21, 19]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode[0])
```

### **Measures of Dispersion**

```python
std_dev = np.std(data)
variance = np.var(data)
range_val = np.max(data) - np.min(data)

print("Standard Deviation:", std_dev)
print("Variance:", variance)
print("Range:", range_val)
```

### **Visualizing Distributions**

```python
sns.histplot(data, kde=True, bins=8)
plt.title("Histogram with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### **Probability Distributions**

```python
# Normal Distribution
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid()
plt.show()
```

### **Skewness and Kurtosis**

```python
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

### **Z-Scores and Outliers**

```python
z_scores = stats.zscore(data)
outliers = np.where(np.abs(z_scores) > 2)

print("Z-scores:", z_scores)
print("Outlier indices:", outliers[0])
```

### **Correlation & Covariance**

```python
# Sample dataset
df = pd.DataFrame({
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Test_Score': [50, 55, 65, 70, 75]
})

correlation = df.corr()
covariance = df.cov()

print("Correlation:\n", correlation)
print("Covariance:\n", covariance)

sns.scatterplot(x='Hours_Studied', y='Test_Score', data=df)
plt.title("Scatterplot: Hours Studied vs Test Score")
plt.show()
```

### **Basic Hypothesis Testing**

```python
# One-sample t-test
sample = [60, 62, 65, 68, 64]
t_stat, p_val = stats.ttest_1samp(sample, 65)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
